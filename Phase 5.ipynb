According to the NBA, "Advanced Stats are a way to study basketball through objective analysis. It is a more in-depth way to look at a simple box score, and more accurately evaluates the skill and production of a player or team". Specifically, there are five attributes being analyzed in this project:

1. **PIE:** Engineered by the NBA in 2013 in order to try and provide an all-in-one stat that summed up a player's performance. In its simplest terms, PIE shows what percentage of game events that a player or team "achieved". An “event” represents stats such as points, assists, and rebounds, as well as personal fouls, turnovers, and defense. PIE is an interesting metric used to gauge a player’s all-around contribution to the game. Almost all statistical categories in the box score are involved in the PIE formula. The formula for PIE calculation is: *(PTS + FGM + FTM – FGA – FTA + Deff.REB + Off.REB/2 + AST + STL + BLK/2 – PF – TO) / (Game.PTS + Game.FGM + Game.FTM – Game.FGA – Game.FTA + Game.Deff.REB + Game.Off.REB/2 + Game.AST + Game.STL + Game.BLK/2 – Game.PF – Game.TO)*. According to the NBA, they introduced this statistic because it incorporates defense into the equation. When a team misses a shot, the PIE rating of all 5 players on the other team goes up. Therefore, we believe that PIE is an adequate variable to measure the performance of teams per season.

**Four Factors:**

2. **Effective Field Goal Percentage (eFG%):** A field goal is a basket scored on any shot taken from the court. The field goal percentage adjusted for the difference in point values for each type of goal (3-pointers vs. 2 pointers). The formula is eFG% = ((Field Goals Made + (0.5 * 3-Pointers Made)) / Field Goals Attempted.


3. **Turnover Percentage (TO Ratio):** A turnover occurs when a team loses possession of the ball to the opposing team before a player takes a shot at their team's basket either from a player getting the ball stolen, stepping out of bounds, having a pass intercepted, committing a violation, or committing an offensive foul. The number of turnovers a player or team averages per 100 possessions and is used to highlight high scorers who tend to not pass the ball. The formula is TO Ratio = (Turnovers 100) / (Field Goals Attempted + (Free Throws Attempted 0.44) + Assists + Turnovers.


4. **Offensive Rebound Percentage (OREB%):** A rebound is a statistic used to describe when a player retrieves a missed field goal or free throw. The offensive rebound percentage is the percentage of offensive rebounds grabbed by a player or team. The formula is OREB% = (Offensive Rebounds) / (Offensive Rebounds + Defensive Rebounds by Opponent).


5. **Free Throw Attempt Rate (FTA Rate):** A free throw is an unhindered shot in basketball made from behind a set line and awarded because of a foul by an opponent. FTA rate is the rate at which free throws are attempted relative to field goals attempted by a player or team. The FTA Rate is used to show how often a player or team goes to the line (**What is the line**), and how good that individual is at drawing fouls. The formula is simply FTA Rate = Free Throws Attempted/Field Goals Attempted.

### Analysis 1

Using our clean dataframes, we can create scatterplots graphing mean PIE rating by win percentage of each team spanning from 1997 to 2021.

```python
# Create scatterplot of all data
sns.regplot(x="pie",y="win%", data = df, line_kws={"color": "red"})

# Fit data into a linear regression
m = LinearRegression().fit(df[['pie']],df['win%'])
plt.xlabel('PIE')
plt.ylabel('Win Percentage')
plt.title('Scatterplot of PIE and Win%')
plt.show()
print("Regression slope is {slope:.2f}".format(slope = m.coef_[0]))
```

```python
predict_accu_test = (m.score(df[['pie']],df[['win%']])*100)
print("Prediction accuracy for test: {:.2f}%".format(predict_accu_test))
```

#### Significance Test: Permutation

```python
# Permutation
def permute(input_array):
    # shuffle is inplace, so copy to preserve input
    permuted = input_array.copy()
    np.random.shuffle(permuted)
    return permuted
```

```python
# how many permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
n_permutations = 10000
permuted_slopes = np.zeros(n_permutations)

for i in range(n_permutations):
    fake_model = LinearRegression().fit(df[['pie']], permute(df['win%']))
    
    permuted_slopes[i] = fake_model.coef_[0]
    
    if np.abs(fake_model.coef_[0]) > np.abs(m.coef_):
        steeper_slopes += 1
```

```python
binsizes, _, _ = plt.hist(permuted_slopes, bins=50)
plt.vlines(m.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.title('Histogram of Permutated Slopes for PIE')
plt.show()
print("Simulated p value: {:}".format(steeper_slopes / n_permutations))
```

**Evaluation of Significance:**

From the regression plot above, we can see that a team's PIE score does have a strong correlation with their win percentage. At a glance, the points on the graph are tightly packed together in a linear fashion and have low residual values. From our regression, we found a high r-squared value of 0.8992, which means that PIE rating accounts for about 89.92% of a change in win percentage. The points are almost perfectly lined up with the regression line and there are very minor outliers which are most apparent on the ends of the graph (around pie scores of 40 and 60). Additionally, the regression coefficient is 4.19, which means that for each percent increase in PIE rating, the average win percentage increases by 4.19%. Although the value appears to be relatively small, the data shows that just an increase in PIE from 50% to 57.5% boosts a team's win percentage from 50% to about 80%. So in reality, every increase in PIE score counts because it will eventually lead to a much higher increase in win percentage. It makes sense that such a small increase in PIE boosts a team's win percentage since a team's PIE typically hovers around 50%. It is much harder, especially in recent years as seen through the decrease in standard deviation from the mythbusting section above, to get to a PIE score of 57.5%, and teams who do so are typically teams that are consistenly outperforming others during a season. To verify that our data did not happen by chance, we ran a permutation test, which yielded us a p value of less than 0.05. In fact, the p value we derived was 0, meaning we can disprove our null hypothesis, and confirm that this data is significant and not a sampling or experimental error.

```python
# Create scatterplot of all data
sns.regplot(x="fta_rate",y="win%", data = df, line_kws={"color": "red"})

# Fit data into a linear regression
m_fta = LinearRegression().fit(df[['fta_rate']],df['win%'])
plt.xlabel('FTA Rate')
plt.ylabel('Win Percentage')
plt.title('Scatterplot of FTA Rate and Win%')
plt.show()
print("Regression slope is {slope:.3f}".format(slope = m_fta.coef_[0]))
```

```python
predict_accu_test = (m_fta.score(df[['fta_rate']],df[['win%']])*100)
print("Prediction accuracy for test: {:.2f}%".format(predict_accu_test))
```

```python
# how many permuted (null hyopthesis) slopes are greater than our observed slope?
steeper_slopes = 0
n_permutations = 10000
permuted_slopes = np.zeros(n_permutations)

for i in range(n_permutations):
    fake_model = LinearRegression().fit(df[['fta_rate']], permute(df['win%']))
    
    permuted_slopes[i] = fake_model.coef_[0]
    
    if np.abs(fake_model.coef_[0]) > np.abs(m_fta.coef_):
        steeper_slopes += 1
```

```python
binsizes, _, _ = plt.hist(permuted_slopes, bins=50)
plt.vlines(m_fta.coef_[0], ymin=0, ymax=np.max(binsizes), color='red')
plt.title('Histogram of Permutated Slopes for FTA Rate')
plt.show()
print("Simulated p value: {:}".format(steeper_slopes / n_permutations))
```

**Evaluation of Significance:**

The scatterplot for FTA Rate, which is the last of the Four Factors in our analysis, also does not show a strong correlation to win percentage. The points plotted are not arranged in a linear fashion and the r-squared value of 0.0179 is insignificantly small as well. That said, the p value is also fairly small, which indicates that this data is significant and we can reject the null hypothesis.

### Bootstrap analysis of each performance metric

Next, we will be performing a bootstrap resampling that generates a 99% confidence interval for the r-coefficients of all five performance metrics (PIE, eFG%, TOV%, OREB%, and FTA Rate). This is being performed so that we can create a boxplot of the bootstrap resampled slopes for each metric as a means to summarize our analysis.

```python
# PIE Bootstrap Resampling
n_bootstrap = 10000
bootstrap_slopes_p = np.zeros(n_bootstrap)

for i in range(n_bootstrap):
    resampled_data = df.sample(n=len(df), replace=True)
    fake_model = LinearRegression().fit(resampled_data[["pie"]], resampled_data["win%"])
    bootstrap_slopes_p[i] = fake_model.coef_[0]
    
c = np.quantile(bootstrap_slopes_p, q=[0.005, 0.995])
print('Simulated CI:', c)
```

### Do Any of These Numbers Even Matter?
Both long-time and new fans can watch a game and pick out what players are good and what players are not. This practice of only watching players to assess their talent level is called the "eye test". There have been efforts at trying to pick out everything to watch for in a player, which the best example for those categories can be found here. The eye test is mainly used to justify that there are factors beyond the stat sheet that also contribute to the success of a player/team. People who believe in the eye-test believe that the best teams with the best talent will win the most games.

The NBA community is split on whether or not to just watch players and assess them or use more advanced analytics to gain a deeper understanding of the game. The main argument against advanced statistics is that the analytics-focused fans bring up meticulous arguments that serve no real relevance when watching NBA games. Using formulas to rank players eventually becomes irrelevant. Stat-padding, which is chasing certain stats to look better on paper, is another argument used against relying on advanced statistics. Players like Russel Westbrook have amazing seasons, but losing records. This same argument can be made for teams as a whole. In the last 20 years, a team other than the top-ranked seed has won the NBA Finals(The championship for the NBA). The epitome of this point can be highlighted in 2016. The Golden State Warriors earned the best record in NBA history, yet they were still unable to win the NBA Finals. Eye-test fanatics would argue that more factors than what showed up on the stat sheet caused them to lose.

Individuals against the eye test argue that there needs to be concrete ways in which to measure a player's performance. The eye test is extremely subjective and opinion-based to an extent. The eye test also creates a limited sample size for analyses because it is nearly impossible to watch a player every minute of every game. Stats are kept of a player's entire performance and eliminate any bias that may come forth.

In The Pattern of Basketball article, this argument between analytics and eve-test is like arguing how to best reinvent the wheel. Since there is no concrete way to quantify the eye-test, this topic will remain a point of disagreement within the NBA community. Ultimately, too much of one thing is never good. Neither of these should be the sole indicator of success but in fact, should actually be a complement to each other. The eye test allows for contextualization of why stats happen. There is a middle-ground where arguers of both sides can sit back and just appreciate the game.
